To simplify rating analysis, the Vote_Average column was grouped into four categories: Blockbuster, Well Received, Averageg Performer, Low Rated, Enabling better comparison and insight generations.

```python
def categorize_col(df, col, labels):
    edges = [df[col].describe()['min'],
             df[col].describe()['25%'],
             df[col].describe()['50%'],
             df[col].describe()['75%'],
             df[col].describe()['max']]
    df[col] = pd.cut(df[col],edges, labels = labels, duplicates = 'drop')
    return df
```

```python
labels = ["Low_Rated", "Average_Performer", "Well_Received", "Blockbuster"]
```

```python
categorize_col(df, 'Vote_Average', labels)
df['Vote_Average'].unique()
```

# Q.1: What is the most frequent genre of movies released on Netflix?

```python
df['Genre'].describe()
```

```python
sns.catplot(y = 'Genre', data = df, kind = 'count',
            order = df['Genre'].value_counts().index,
            color = '#f483d7')
plt.title('Genre Column Distribution')
plt.show()
```

# Q.2: Which has highest votes in vote average columns??

```python
sns.catplot(y = 'Vote_Average', data = df, kind = 'count',
            order = df['Vote_Average'].value_counts().index,
            color = '#4beb9e')
plt.title('Vote Distribution')
plt.show()
```
